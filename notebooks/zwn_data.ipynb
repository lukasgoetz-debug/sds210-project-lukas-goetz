```python
# 1. Import of packages

import pandas as pd
import geopandas as gpd
from shapely import wkt

# 2. Read Zueriwieneu data

fp = "../data/raw/stzh.zwn_meldungen_p.csv"
zh_df = pd.read_csv(fp)

print("Data loaded successfully")

# 3. Cleaning and preparing the csv-data of Zueriwieneu requests with Pandas and convert it to a spatial GeoDataFrame

# Deleting unnecessary columns
zh_df = zh_df.drop(columns=["objectid", "agency_sent_datetime", "service_code", "media_url"],
    errors="ignore"
)

# Parse the date-time columns
zh_df["requested_datetime"] = pd.to_datetime(
    zh_df["requested_datetime"],
    format="%Y-%m-%dT%H:%M:%S"
)

zh_df["updated_datetime"] = pd.to_datetime(
    zh_df["updated_datetime"],
    format="%Y-%m-%dT%H:%M:%S"
)

# Check for null values
display(zh_df.isna().sum().sort_values(ascending=False))

# Fill the shown missing values for the respective columns
zh_df["title"] = zh_df["title"].fillna("")
zh_df["detail"] = zh_df["detail"].fillna("")
zh_df["service_notice"] = zh_df["service_notice"].fillna("")

# Export the cleaned csv to work with it later

# Define the output file name
output_csv = "../data/cleaned/stzh.zwn_meldungen_p_cleaned.csv"

# Save DataFrame to a new CSV file
zh_df.to_csv(output_csv, index=False)

# Convert the standard DataFrame into a GeoDataFrame
zh_df["geometry"] = zh_df["geometry"].apply(wkt.loads)

zh_gdf = gpd.GeoDataFrame(
    zh_df, 
    geometry="geometry",
    crs="EPSG:2056"
)

# Export the converted gdf as .gpkg

# Define output file name
output_gpkg = "../data/cleaned/stzh.zwn_meldungen_p_cleaned.gpkg"

# Save as GeoPackage
zh_gdf.to_file(
    output_gpkg,
    layer="meldungen",
    driver="GPKG"
)

# Show head and info to check changes
display(zh_gdf.head())
zh_gdf.info()
```
