```python
# 1. Import of packages

import pandas as pd
import geopandas as gpd
from shapely import wkt
```

```python
# 2. Read population data excel and arrange 2013 - 2025 total population for 'Quartiere' to one csv worksheet

fp = "../data/raw/BEV321T3211_auslaendische-Wohnbevoelkerung_Bevoelkerung_nach-Herkunft-Stadtkreis-Stadtquartier.xlsx"

# Years to include
years = range(2013, 2026)

# Empty list to store yearly data
all_years = []

for year in years:

    # Read one sheet (= one year)
    year_df = pd.read_excel(
        fp,
        sheet_name=str(year),
        skiprows=8
    )

    # Rename first column
    year_df = year_df.rename(columns={"Unnamed: 0": "quartier"})

    # Keep only relevant columns
    year_df = year_df[["quartier", "Total"]]

    # Remove "Ganze Stadt"
    year_df = year_df[year_df["quartier"] != "Ganze Stadt"]

    # Remove "Kreis" rows ("~" reverses True/False)
    year_df = year_df[~year_df["quartier"].str.startswith("Kreis")]

    # Rename "Total" column to the corresponding year
    year_df = year_df.rename(columns={"Total": str(year)})

    # Store cleaned dataframe
    all_years.append(year_df)

# Merge all years together
pop_df = all_years[0]

for df in all_years[1:]:
    pop_df = pop_df.merge(df, on="quartier")

# Save as CSV
output_file = "../data/cleaned/quartier_population_2013_2025.csv"

pop_df.to_csv(output_file, index=False)

print("CSV successfully created!")

display(pop_df.head())

print("Data loaded successfully")

display(pop_df.head())
pop_df.info()
```
