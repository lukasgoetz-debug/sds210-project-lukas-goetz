```python
# 1. Import of packages

import pandas as pd
import geopandas as gpd
from shapely import wkt
```

```python
# 2. Read Zueriwieneu data

fp = "data/stzh.zwn_meldungen_p.csv"
zh_df = pd.read_csv(fp)

print("Data loaded successfully")
```

```python
# 3. Cleaning and preparing the csv-data of Zueriwieneu requests with Pandas

# Deleting unnecessary columns
zh_df = zh_df.drop(columns=["objectid", "agency_sent_datetime", "service_code", "media_url"])

# Setting service_request_id as the index of the df
# (object_id is not stable over time according to the data description on the web page)
zh_df = zh_df.set_index("service_request_id")

# Parse the date-time columns
zh_df["requested_datetime"] = pd.to_datetime(
    zh_df["requested_datetime"],
    format="%Y-%m-%dT%H:%M:%S"
)

zh_df["updated_datetime"] = pd.to_datetime(
    zh_df["updated_datetime"],
    format="%Y-%m-%dT%H:%M:%S"
)

# Export the cleaned csv to work with it later

# Define the output file name
output_file = "data/stzh.zwn_meldungen_p_cleaned.csv"

# Save DataFrame to a new CSV file
zh_df.to_csv(output_file, index=False)

# Convert the standard DataFrame into a GeoDataFrame
zh_df["geometry"] = zh_df["geometry"].apply(wkt.loads)

zh_gdf = gpd.GeoDataFrame(
    zh_df, 
    geometry="geometry",
    crs="EPSG:2056"
)

# Show head and info to check changes
display(zh_df.head())
zh_df.info()
```
